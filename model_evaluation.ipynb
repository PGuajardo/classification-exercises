```python
model_precision = (subset.actual == subset.model2).mean()
print(f'model2 precision: {model_precision:.2%}')

# Baseline Precision: # only look at the subset where we made positive prediction (i.e. model1 == 'Defect')

subset = c3[c3.baseline_prediction == 'Defect']
baseline_precision = (subset.actual == subset.model2).mean()
print(f'baseline precision: {baseline_precision:.2%}')
```
